<br>
برای رفع این مشکل باید

<br>و
3.	‏اختصاص لیبل -100 به توکن‌[<ها]ی s[> و ]</s>. زیرا مقدار لیبل منفی توسط تابع loss PyTorch نادیده گرفته می‌شود.


## توضیح کد:
<div dir="rtl" style="font-family: Vazir; width: 85%; font-size: 16px;">
برای alignment تابع tokenize_and_align_labels تعریف شده‌است. در این تابع ابتدا توکن‌ها در به ازای هر نمونه مجموعه داده که به صورت کلمه هستند با استفاده از توکنایزر XLMRobert و با قرار دادن مقدار True برای is_split_into_words به sub word تقسیم می‌شوند، همانطور که قبلا بیان شد توکنایزر XLMRobert دو توکن جدید > و </s> را به ‌توکن‌های موجود بعد از توکنایز کردن اضافه می‌کند. با تعریف یک حلقه for بر روی مقادیر ستون ner_tags به ازای هر نمونه‌ ابتدا به کمک متد word_ids آی دی کلمات معادل هر کدام از sub word ها تعریف می‌شود. سپس متغیر previous_word_idx ( برای اینکه تنها به اولین توکن sub word کلمه لیبل کلمه اختصاص داده ‌شود، استفاده خواهد شد) و لیست label_ids (برای ذخیره لیبل‌های جدید برای توکن‌های جدید به ازای هر نمونه)  تعریف می‌شوند.
<br>
با iterate بر روی word_ids توکن‌ها، اگر مقدار word_id توکن برابر با None باشد یعنی آن توکن یکی از دو توکن جدید اضافه شده توسط توکنایزر است، پس لیبل -100 را خواهد گرفت. اگر word_id آن توکن برابر با word_id توکن قبلی نباشد یعنی دو توکن متعلق به یک کلمه نیستند، پس لیبل آن کلمه به عنوان لیبل آن توکن ذخیره خواهد شد. در نهایت اگر word_id آن توکن برابر با word_id توکن قبلی باشد یعنی هر دو توکن متعلق به یک کلمه هستند و از آنجا که قرار است تنها توکن اول کلمه لیبل کلمه را بگیرد، این توکن دیگر کلمه لیبل -100 خواهد گرفت. در نهایت ner_tags جدید برای آن نمونه در لیست labels ذخیره می‌شود. این تابع با استفاده از متد map بر روی سه دیتاست train، test و validation اعمال می‌شود.
<br>
همچنین با استفاده از ‏‏Data Collator‏‏، batch ای از نمونه ها ایجاد می‌شود تا تعداد توکن‌های هر نمونه را به صورت ‏‏پویا‏‏ به طولانی ترین طول تعداد توکن‌ها در یک batch برساند، به جای اینکه کل دیتاست را به  اندازه حداکثر طول تعداد توکن‌های pad کند.‏

```python
from transformers import XLMRobertaTokenizerFast

tokenizer = XLMRobertaTokenizerFast.from_pretrained("xlm-roberta-base")

def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)

    labels = []
    for i, label in enumerate(examples[f"ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                label_ids.append(label[word_idx])
            else:
                label_ids.append(-100)
            previous_word_idx = word_idx
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

```python
train_dataset = train_dataset.map(tokenize_and_align_labels, batched=True)
val_dataset = val_dataset.map(tokenize_and_align_labels, batched=True)
test_dataset = test_dataset.map(tokenize_and_align_labels, batched=True)
```

#### توضیح کد:
<div dir="rtl" style="font-family: Vazir; width: 85%; font-size: 16px;">
برای TokenClassifiaction با استفاده از xlmRoberta کلاس xlmRobertaSequenceClassification تعریف شده‌است. در این کلاس ابتدا لایه‌های مدل که شامل xlmRoberta،  dropout (برای جلوگیری از overfiting) و linear (برای طبقه‌بندی) که سایز ورودی آن به اندازه سایز آخرین لایه hidden مدل xlmRoberta  و سایز خروجی به اندازه تعداد لیبل‌ها است، هستند، تعریف می‌شوند.
<br>
در قسمت forward، id توکن‌ها و attention mask به عنوان ورودی به مدل xlmRoberta داده می‌شوند و آخرین لایه پنهان مدل به عنوان خروجی تعریف می‌شود. سپس خروجی به لایه dropout برای جلوگیری از overfitting داده می‌شود و خروجی این لایه به لایه linear برای طبقه بندی و تولید logit ها داده می‌شوند
<br>.
برای محاسبه loss اگر لیبل وجود داشته باشد ابتدا تابع loss تعریف می‌شود. سپس ایندکس attention mask هایی که مقداری برابر با 1 دارند (pad نیستند) ذخیره می‌شود تا براساس آن loss برای مقادیر غیر pad محاسبه شود. در گام بعد logit ها و لیبل‌های active (غیر pad) برای محاسبه loss، flat  می‌شوند و loss محاسبه می‌شود. در نهایت باتوجه به اینکه مقدار gradient_checkpointing در تعریف argument های آموزش مدل True تعریف می‌شود، در این کلاس gradient checkpointing برای مدل فعال می‌شو
<br>د.
در بخش بعد TrainingArguments مشابه آنچه در صورت سوال داده شده تعریف می‌ش
<br>
برای محاسبه معیارهای خواسته شده متریک seqeval به کمک کتابخانه evaluate بارگذاری می‌شود. سپس تابع compute_metrics تعریف شده که ورودی‌های آن پیش‌بینی مدل و لیبل‌های اصلی ‌توکن‌ها (به صورت id) است. پیش‌بینی مدل به صورت logit هستند که  با استفاده از تابع argmax به id لیبل تبدیل می‌شوند. با iterate بر روی پیش‌بینی مدل و لیبل‌های اصلی ‌توکن‌ها بررسی می‌شود اگر مقدار لیبل برابر با -s>) نباشد، id ها به لیبل با استفاده از دیکشنری id2label که در بخش قبل ایجاد شد تبدیل می‌شوند. سپس به کمک لیبل‌های ایجاد شده معیارهای خواسته شده محاسبه و برگردانده می‌شوند.
<br>
در نهایت trainer برای آموزش مدل با تعریف پارامترها شامل مدل، آرگومان‌های آموزش، داده‌های آموزش و ارزیابی، compute_metrics و data_collator تعریف می‌شود و مدل آموزش داده می‌شود.ود.


```python
 #write your code here
import torch
import torch.nn as nn
from transformers import XLMRobertaModel


class xlmRobertaSequenceClassification(nn.Module):
    def __init__(self, num_labels):
        super(xlmRobertaSequenceClassification, self).__init__()

        self.xlmRoberta = XLMRobertaModel.from_pretrained("xlm-roberta-base")
        self.dropout = nn.Dropout(self.xlmRoberta.config.hidden_dropout_prob)
        self.linear = nn.Linear(self.xlmRoberta.config.hidden_size, num_labels)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.xlmRoberta(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = outputs.last_hidden_state

        sequence_output = self.dropout(sequence_output)
        logits = self.linear(sequence_output)

        loss = None
        if labels is not None:
            loss_fct = nn.CrossEntropyLoss()
            active_loss = attention_mask.view(-1) == 1
            active_logits = logits.view(-1, logits.shape[-1])[active_loss]
            active_labels = labels.view(-1)[active_loss]
            loss = loss_fct(active_logits, active_labels)

        return {'loss': loss, 'logits': logits} if loss is not None else logits

    def gradient_checkpointing_enable(self, **kwargs):
        self.xlmRoberta.gradient_checkpointing_enable()


model = xlmRobertaSequenceClassification(num_labels=21)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
output_dir="Roberta-fa-en-ner",
report_to="none", log_level="error",
num_train_epochs=1,
gradient_checkpointing=True,
eval_accumulation_steps=10,
per_device_train_batch_size=24,
per_device_eval_batch_size=24,
seed=42,
logging_strategy="steps", eval_strategy="steps",
save_steps=1e6, weight_decay=0.01, disable_tqdm=False,
logging_steps=1e6, eval_steps=400, push_to_hub=False)
```

```python
import evaluate
import numpy as np

metric = evaluate.load("seqeval")

def compute_metrics(eval_pred):
    predictions, references = eval_pred

    predictions = np.argmax(predictions, axis=2)

    filtered_predictions = []
    filtered_references = []

    for preds, refs in zip(predictions, references):
        filtered_preds = []
        filtered_refs = []
        for p, r in zip(preds, refs):
            if r != -100:
                filtered_preds.append(id2label[p])
                filtered_refs.append(id2label[r])
        filtered_predictions.append(filtered_preds)
        filtered_references.append(filtered_refs)

    results = metric.compute(predictions=filtered_predictions, references=filtered_references)

    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
    data_collator=data_collator
)
```

```python
trainer.train()
```

```python
#write your code here

from datasets import load_dataset, DatasetDict

dataset = load_dataset("cnn_dailymail", "3.0.0")

print(dataset)

train_dataset = dataset["train"].shuffle(seed=42).select(range(5000))
val_dataset = dataset["validation"].shuffle(seed=42).select(range(500))
test_dataset = dataset["test"].shuffle(seed=42).select(range(100))

dataset_dict = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset,
    "test": test_dataset
})


print(dataset_dict)
```

```python
print("article:", dataset_dict["train"][0]["article"])
print("summary:", dataset_dict["train"][0]["highlights"])
```

```python
def calculate_lengths(dataset, text_column, summary_column):
    text_lengths = [len(sample[text_column].split()) for sample in dataset]
    summary_lengths = [len(sample[summary_column].split()) for sample in dataset]

    min_text_length = min(text_lengths)
    max_text_length = max(text_lengths)
    avg_text_length = sum(text_lengths) / len(text_lengths)

    min_summary_length = min(summary_lengths)
    max_summary_length = max(summary_lengths)
    avg_summary_length = sum(summary_lengths) / len(summary_lengths)

    return (min_text_length, max_text_length, avg_text_length), (min_summary_length, max_summary_length, avg_summary_length)

text_stats, summary_stats = calculate_lengths(dataset_dict["train"], "article", "highlights")

print("article- min:", text_stats[0], "max:", text_stats[1], "mean:", text_stats[2])
print("summary- min:", summary_stats[0], "max:", summary_stats[1], "mean:", summary_stats[2])
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype="float16"
)


model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B-Instruct", quantization_config=quantization_config)

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B-Instruct")

if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '<pad>'})
    model.resize_token_embeddings(len(tokenizer))
```

مفهوم 4:
<br>
برای دستیابی به مدل‌های زبانی کوچک با دقت مناسب و بهینه‌سازی عملکرد آن‌ها با منابع محاسباتی محدود، تکنیک‌های متنوعی وجود دارد. این تکنیک‌ها شامل بهینه‌سازی معماری، روش‌های آموزشی، و فشرده‌سازی مدل‌ها می‌شوند. که این تکنیک‌ها را به شرح زیر هستند:
    <br>
1. فشرده‌سازی مدل از طریق Distillation: فرآیندی است که طی آن مدل کوچک‌تر (Student Model) با یادگیری از پیش‌بینی‌های مدل بزرگ‌تر و پیشرفته‌تر (Teacher Model) آموزش می‌بیند. در این فرآیند، خروجی مدل بزرگ‌تر به عنوان هدف برای مدل کوچک‌تر استفاده می‌شود که کمک می‌کند مدل کوچک‌تر اطلاعات بیشتری را نسبت به آموزش با داده خام کسب کند.
    <br>
2. هرس کردن (Pruning ): شامل حذف وزن‌ها یا نودهای کم‌اهمیت از شبکه عصبی است. این فرآیند مدل را فشرده‌تر و سریع‌تر می‌کند، اما باید با دقت انجام شود تا کاهش دقت به حداقل برسد.
    <br>
3. Quantization : شامل کاهش تعداد بیت‌های مورد استفاده برای نمایش وزن‌ها است. به عنوان مثال، کاهش دقت از 32 بیت به 8 بیت (INT8) یا 16 بیت. این تکنیک کاهش اندازه مدل و سرعت بخشیدن به محاسبات را با کاهش حداقلی دقت ممکن می‌سازد.
    <br>
4.  Sparse Training : در این تکنیک، مدل‌ها با وزن‌های پراکنده آموزش داده می‌شوند، به طوری که بسیاری از وزن‌ها صفر باقی می‌مانند. این موضوع به کاهش اندازه مدل کمک می‌کند و پردازش سریع‌تر می‌شود.
    <br>
5. Parameter Sharing : اشتراک‌گذاری پارامترها بین بخش‌های مختلف شبکه کمک می‌کند تعداد پارامترهای مستقل کاهش یابد و مدل کوچک‌تر شود.
    <br>
6. Embedding Layer Factorization : در مدل‌های زبانی که از Embedding کلمه استفاده می‌کنند، کاهش اندازه لایه‌های Embedding با استفاده از فاکتورگیری ماتریس یا فشرده‌سازی بردارها امکان‌پذیر است.
    <br>
7.  Transfer Learning: استفاده از Transfer Learning به مدل‌های کوچک اجازه می‌دهد از دانشی که در مدل‌های بزرگ‌تر به دست آمده بهره ببرند. این تکنیک به خصوص در وظایف خاص (domain-specific) که داده‌های کمی وجود دارد، مؤثر است.
    <br>
8.  Head Pruning در مدل‌های ترنسفورمر: شامل حذف head های بی‌اهمیت در لایه‌های Attention در مدل‌های ترنسفورمر است. این تکنیک اندازه مدل و پیچیدگی محاسباتی را کاهش می‌دهد و در عین حال عملکرد نسبی آن حفظ می‌شود.
