NER جزء استخراج اطلاعات است که هدف آن شناسایی و طبقه بندی موجودیت های نامگذاری شده در متن بدون ساختار است. NER شامل شناسایی اطلاعات کلیدی در متن و طبقه بندی به مجموعه ای از دسته های از پیش تعریف شده است. موجودیت چیزی است که به طور مداوم در مورد آن صحبت می شود یا در متن به آن اشاره می شود، مانند نام افراد، سازمان ها، مکان ها، عبارات زمانی، کمیت ها، درصدها و دسته های از پیش تعریف شده بیشتر.
<br>
    <br>
هدف اصلی NER، استخراج اطلاعات ساخت یافته از متن های غیرساخت یافته است. به عبارت دیگر، NER به سیستم ها کمک می کند تا اطلاعات مهم و قابل استخراج از متن را شناسایی کرده و آن ها را به صورت ساخت یافته نمایش دهند. این اطلاعات می تواند برای طیف گسترده ای از کاربردها، از جمله جستجوی اطلاعات، خلاصه سازی متن، ترجمه ماشینی، تحلیل احساسات و پاسخگویی به سوالات استفاده شود.

```python
import re

url_pattern = r'(https?:\/\/(?:www\.[a-zA-Z0-9-]+\.[^\s]{2,}))'
email_pattern = r'([a-zA-Z0-9]+\.?[^\s][a-zA-Z0-9]+\@[a-zA-Z0-9-]+\.[^\s]{2,})'
phone_num_pattern = r'(\+[0-9]+[-|\s][0-9]+[-|\s][a-zA-Z0-9]+[-|\s][a-zA-Z0-9]+)'

url = []
email = []
phone_num = []

with open('ner.txt', 'r') as file:
    
    for line in file:
        
        url_matches = re.findall(url_pattern, line)
        email_matches = re.findall(email_pattern, line)
        phone_num_matches = re.findall(phone_num_pattern, line)
        
        if url_matches:
            url.extend(url_matches)
            
        if email_matches:
            email.extend(email_matches)
            
        if phone_num_matches:
            phone_num.extend(phone_num_matches)
            

print('URLs:', url)
print('\nemails:', email)
print('\nPhone numbers:', phone_num)
```

الگوریتم های Levenshtein distance و Damerau-Levenshtein distance برای اندازه گیری فاصله و تفاوت بین دو رشته استفاده می شوند. این الگوریتم ها برای تشخیص شباهت یا تفاوت میان کلمات، جملات یا رشته های متنی مورد استفاده قرار می گیرند و کاربردهای متنوعی در پردازش زبان طبیعی، بررسی غلط املایی و جستجو دارند.
<br> 
الگوریتم Levenshtein distance (edit distance) میزان حداقل تعداد عملیات لازم برای تبدیل یک رشته به رشته دیگر را محاسبه می کند.  این عملیات ها شامل حذف  (deletion)، درج (insertion) و جایگزینی (substitution) یک کاراکتر هستند. برای محاسبه این فاصله از الگوریتم برنامه‌نویسی پویا استفاده می‌شود. این الگوریتم یک ماتریس ایجاد می‌کند و طول دو رشته را به عنوان ابعاد در نظر می‌گیرد. با پر کردن سلول‌های این ماتریس به صورت بازگشتی بهینه‌ترین راه برای تبدیل یک رشته به رشته دیگر پیدا می‌شود. ماتریس شامل مقدارهایی است که هر مقدار نشان می‌دهد که چند عملیات لازم است تا قسمتی از رشته اول به قسمتی از رشته دوم تبدیل شود. 
 <br>
الگوریتم Damerau-Levenshtein distance علاوه بر عملیات حذف، درج و جایگزینی، عملیات جابه‌جایی دو کاراکتر مجاور (transposition)  را نیز در نظر می‌گیرد. مانند  Levenshtein distance، این الگوریتم نیز از یک ماتریس پویا استفاده می‌کند. تفاوت در این است که در این الگوریتم، علاوه بر محاسبه عملیات جایگزینی، درج و حذف، الگوریتم بررسی می‌کند که آیا می‌توان با یک عملیات جابجایی، به نتیجه مطلوب رسید یا خیر. اگر جابجایی دو کاراکتر مجاور امکان‌پذیر باشد و منجر به کمترین تعداد عملیات شود، از این امکان نیز استفاده می‌شود.

```python
import numpy as np

def Levenshtein(s1, s2):
    
    m = len(s1)
    n = len(s2)
    d = np.zeros((m+1, n+1))
    
    for i in range(m+1):
        d[i, 0] = i
        
    for j in range(n+1):
        d[0, j] = j
    
    for i in range(1, m+1):
        for j in range(1, n+1):
            
            if s1[i-1] == s2[j-1]:
                cost = 0
            else:
                cost = 2
            
            d[i, j] = min(d[i - 1, j] + 1,           # Deletion
                           d[i, j - 1] + 1,          # Insertion
                           d[i - 1, j - 1] + cost)   # substitution
                
    return d[m, n]


def Damerau_Levenshtein(s1, s2):
    
    m = len(s1)
    n = len(s2)
    d = np.zeros((m+1, n+1))
    
    for i in range(m+1):
        d[i, 0] = i
        
    for j in range(n+1):
        d[0, j] = j
    
    for i in range(1, m+1):
        for j in range(1, n+1):
            
            if s1[i-1] == s2[j-1]:
                cost = 0
            else:
                cost = 2
            
            d[i, j] = min(d[i - 1, j] + 1,           # Deletion
                           d[i, j - 1] + 1,          # Insertion
                           d[i - 1, j - 1] + cost)   # substitution
            
            if i > 1 and j > 1 and s1[i - 1] == s2[j - 2] and s1[i - 2] == s2[j - 1]:     #transposition
                d[i, j] = min(d[i, j], d[i - 2, j - 2] + 1)
                
    return d[m, n]
                

print(f"'kitten' and 'sitting': Levenshtein distance is {Levenshtein('kitten', 'sitting')} and Damerau-Levenshtein distance is {Damerau_Levenshtein('kitten', 'sitting')}")
print(f"'saturday' and 'sunday': Levenshtein distance is {Levenshtein('saturday', 'sunday')} and Damerau-Levenshtein distance is {Damerau_Levenshtein('saturday', 'sunday')}")
print(f"'book' and 'back': Levenshtein distance is {Levenshtein('book', 'back')} and Damerau-Levenshtein distance is {Damerau_Levenshtein('book', 'back')}")
print(f"'algorithm' and 'logarithm': Levenshtein distance is {Levenshtein('algorithm', 'logarithm')} and Damerau-Levenshtein distance is {Damerau_Levenshtein('algorithm', 'logarithm')}")
print(f"'' and 'test': Levenshtein distance is {Levenshtein('', 'test')} and Damerau-Levenshtein distance is {Damerau_Levenshtein('', 'test')}")
print(f"'abc' and 'acb': Levenshtein distance is {Levenshtein('abc', 'acb')} and Damerau-Levenshtein distance is {Damerau_Levenshtein('abc', 'acb')}")
```

برای محاسبه فاصله Levenshtein تابع Levenshtein تعریف شده‌است که وردی‌های آن دو کلمه موردنظر هستند. سپس آرایه دو بعدی d برای ذخیره هزینه‌ها با مقدار اولیه صفر، که اندازه ابعاد آن برابر با طول دو کلمه است تعریف شده‌است. باتوجه به الگوریتم Levenshtein مقدار سطرهای ستون صفر با مقدار ایندکس آن سطر و مقدار ستون‌های سطر صفر با ایندکس آن ستون جایگزین می‌شوند. برای مقداردهی سایر درایه‌ها، ابتدا هزینه جایگزینی تعیین می‌شود، اگر دو حرف قبلی با هم برابر باشند هزینه جایگزینی برابر با صفر و در غیر اینصورت برابر با 2 می‌شود. سپس مقدار آن داریه برابر با مینیمم مقادیر ناشی از insert، delete و substitute خواهد بود.

برای محاسبه فاصله Damerau_Levenshtein، تابع Damerau_Levenshtein تعریف شده‌است که تقریبا مشابه تابع Levenshtein است و تنها تفاوت آنها در اضافه شدن جابجایی است. برای جابجایی 4 شرط بررسی می‌شود. دو شرط اول بررسی می‌کنند که کاراکترهای موردنظر، کاراکترهای ابتدایی دو کلمه نباشند. و دو شرط دیگر برابری کاراکترهای دو رشته برای جابجایی را بررسی می‌کند. در صورت برقراری 4 شرط هزینه محاسبه می‌شود.

<div dir="rtl" style="font-family: Vazir; width: 85%; font-size: 16px;">

با افزایش n، تعداد ترکیبات ممکن از کلمات یا کاراکترها به صورت نمایی افزایش می‌یابد. این امر باعث می‌شود که محاسبات پیچیده‌تر و زمان‌برتر شوند.
    <br>
با افزایش طول n-gram ها، احتمال مشاهده هر n-gram خاص به کاهش می‌یابد. این به این معنی است که بسیاری از n-gram ها ممکن است تنها یک یا چند بار در متن ظاهر شوند، که این امر می‌تواند منجر به تخمین‌های نادرست از احتمال وقوع آن‌ها شود.
  <br>  
برای آموزش مدل‌های زبانی مبتنی بر n-gram، به مقدار زیادی داده نیاز داریم. با افزایش n، این نیاز به داده بیشتر می‌شود.
<br>    
مدل‌های n-gram بزرگتر ممکن است به جای یادگیری الگوهای کلی، به جزئیات خاص داده‌های آموزشی وابسته شوند. این امر می‌تواند باعث کاهش توانایی مدل در تعمیم به داده‌های جدید شود.


```python
from collections import defaultdict
import random

class NGram():
    def __init__(self, n):
        self.n = n
        self.ngrams = defaultdict(int)     
        self.n_1grams = defaultdict(int)   
        self.probabilities = defaultdict(float)

    def generate_ngram(self, tokens):
        
        for i in range(len(tokens) - self.n + 1):
            
            ngram = tuple(tokens[i: i + self.n]) 
            self.ngrams[ngram] += 1

            n_1gram = tuple(tokens[i: i + self.n - 1])
            self.n_1grams[n_1gram] += 1

            
    def train(self, tokens):
        
        self.generate_ngram(tokens)
        
        for ngram, count in self.ngrams.items():
            n_1gram = ngram[:-1]
            self.probabilities[ngram] = count / self.n_1grams[n_1gram]

            
    def predict_next_word(self, context):
        
        context_tuple = tuple(context[-(self.n - 1):])
        candidates = {}

        for ngram, prob in self.probabilities.items():
            if ngram[:-1] == context_tuple:
                candidates[ngram[-1]] = prob 

        next_word = max(candidates, key=candidates.get)
        
        return next_word

    def generate_sentence(self, start_tokens, max_length):
       
        sentence = start_tokens
        
        for _ in range(max_length - len(start_tokens)):
            
            next_word = self.predict_next_word(sentence)
            sentence.append(next_word)
            
        return ' '.join(sentence)
```

```python
# tokenization 

tokens = []

for l in read_corpus("ferdowsi.txt"):
    tokens.extend(tokenizer.encode(l).tokens)
```

```python
# 2 gram

ngram_model2 = NGram(n=2)

ngram_model2.train(tokens)

sentence = ngram_model2.generate_sentence(tokens[0:2], max_length=200)
print(f"Generated sentence: {sentence}")
```

Perplexity یک معیارارزیابی است که برای بررسی کیفیت پیش‌بینی‌های مدل‌های زبانی استفاده می‌شود.perplexity  نشان می‌دهد که یک مدل زبان چقدر در پیش‌بینی توالی کلمات در یک متن خوب عمل می‌کند.
    <br>
perplexity  یک مدل زبانی روی یک مجموعه test، معکوس احتمال آن مجموعه test (یک تقسیم بر احتمال مجموعه test) است که به ازای تعداد کلمات (یا توکن‌ها) نرمال‌سازی شده است.<br>
 اگر perplexity یک متن روی یک مدل زبانی کم باشد، نشان‌دهنده این است که مدل توانسته است به خوبی توالی کلمات را پیش‌بینی کند. یعنی احتمال بالایی به توالی کلمات داده است و در نتیجه، مدل به خوبی آموزش دیده و قادر به تعمیم به داده‌های جدید است.


```python
#write your code here
import math

def calculate_perplexity(tokens, probability, n):
    
        total_log_prob = 0
        N = len(tokens)
        
        for i in range(len(tokens) - n + 1):
            
            ngram = tuple(tokens[i: i + n]) 
            
            if probability[ngram]:
                total_log_prob += math.log(probability[ngram])
            else:
                total_log_prob += math.log(1e-6)
                
        
        avg_log_prob = total_log_prob / N
        perplexity = math.exp(-avg_log_prob)
        
        return perplexity
```

```python
from collections import defaultdict

class NGramClassifier:
    def __init__(self, n, classes):
        self.n = n
        self.class_ngrams = defaultdict(lambda: defaultdict(int))
        self.classes = classes

    def generate_ngrams(self, tokens):
        
        ngrams = []
        for i in range(len(tokens) - self.n + 1):
            ngrams.append(tuple(tokens[i: i + self.n]))
            
        return ngrams


    def train(self, sentences, labels):

        for tokens, label in zip(sentences, labels):
            ngrams = self.generate_ngrams(tokens)
            
            for ngram in ngrams:
                self.class_ngrams[label][ngram] += 1

    def predict(self, tokens):

        ngrams = self.generate_ngrams(tokens)
        class_scores = {cls: 0 for cls in self.classes}

        for ngram in ngrams:
            for cls in self.classes:
                class_scores[cls] += self.class_ngrams[cls].get(ngram, 0)

        return max(class_scores, key=class_scores.get)

    
    def evaluate(self, sentences, labels):

        predictions = [self.predict(sentence) for sentence in sentences]

        true_positives = defaultdict(int)
        false_positives = defaultdict(int)
        false_negatives = defaultdict(int)

        for predicted_label, actual_label in zip(predictions, labels):
            if predicted_label == actual_label:
                true_positives[predicted_label] += 1
            else:
                false_positives[predicted_label] += 1
                false_negatives[actual_label] += 1

        # macro ave.
        precision_avg = sum(true_positives[cls] / (true_positives[cls] + false_positives[cls]) if true_positives[cls] + false_positives[cls] > 0 else 0 for cls in self.classes) / len(self.classes)
        recall_avg = sum(true_positives[cls] / (true_positives[cls] + false_negatives[cls]) if true_positives[cls] + false_negatives[cls] > 0 else 0 for cls in self.classes) / len(self.classes)

        accuracy = sum(true_positives.values()) / len(labels)

        return accuracy, precision_avg, recall_avg
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(tokenized_dataset, df['Suggestion'], test_size=0.3, random_state=42)
```

```python
ngram_2_classifier = NGramClassifier(n=2, classes = set(y_train))
ngram_2_classifier.train(X_train, y_train)

accuracy_2, precision_2, recall_2 = ngram_2_classifier.evaluate(X_test, y_test)
print(f"2-Gram Model - Accuracy: {accuracy_2:.2f}, Precision: {precision_2:.2f}, Recall: {recall_2:.2f}")
```

```python
ngram_3_classifier = NGramClassifier(n=3, classes = set(y_train))
ngram_3_classifier.train(X_train, y_train)

accuracy_3, precision_3, recall_3 = ngram_3_classifier.evaluate(X_test, y_test)
print(f"3-Gram Model - Accuracy: {accuracy_3:.2f}, Precision: {precision_3:.2f}, Recall: {recall_3:.2f}")
```

```python
from keras.datasets import mnist
import numpy as np
from sklearn.cluster import KMeans

(x_train, y_train), (x_test, y_test) = mnist.load_data()

# preprocessing
x_train = x_train / 255.0
x_test = x_test / 255.0


# training tokenizer
def extract_patches(image, size_patch, stride):
    patches = []
    for i in range(0, image.shape[0] - size_patch + 1, stride):
        for j in range(0, image.shape[1] - size_patch + 1, stride):
            patch = image[i:i + size_patch, j:j + size_patch]
            patches.append(patch.flatten()) 
    return np.array(patches)


size_patch = 9
stride = 9
n_clusters = 256


all_patches = []
all_patches = np.concatenate([extract_patches(image, size_patch, stride) for image in x_train])
all_patches = np.array(all_patches)

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(all_patches)
```

```python
# tokenization
def tokenize_image(image, size_patch, stride, kmeans):
    patches = extract_patches(image, size_patch, stride)
    clusters = kmeans.predict(patches)
    return clusters

print("Tokens example: ", tokenize_image(x_test[0], 9, 9, kmeans))

train_tokens = [tokenize_image(image, size_patch, stride, kmeans) for image in x_train]
test_tokens = [tokenize_image(image, size_patch, stride, kmeans) for image in x_test]
```

```python
#write your code here
from collections import defaultdict
from sklearn.metrics import accuracy_score, precision_score, recall_score

class NGramClassifier:
    def __init__(self, n, classes):
        self.n = n
        self.class_ngrams = defaultdict(lambda: defaultdict(int))
        self.classes = set()

    def generate_ngrams(self, tokens):

        ngrams = [tuple(tokens[i:i+self.n]) for i in range(len(tokens)-self.n+1)]
        return ngrams



    def train(self, sentences, labels):

        for tokens, label in zip(sentences, labels):
            ngrams = self.generate_ngrams(tokens)
            self.classes.add(label)
            for ngram in ngrams:
                self.class_ngrams[label][ngram] += 1
                

    def predict(self, tokens):

        ngrams = self.generate_ngrams(tokens)
        class_scores = {cls: 0 for cls in self.classes}

        for ngram in ngrams:
            for cls in self.classes:
                class_scores[cls] += self.class_ngrams[cls].get(ngram, 0)

        return max(class_scores, key=class_scores.get)

    
    def evaluate(self, sentences, labels):

        predictions = [self.predict(sentence) for sentence in sentences]
        accuracy = accuracy_score(labels, predictions)
        return accuracy

```

```python
ngram_classifier = NGramClassifier(n=4, classes = set(y_train))
ngram_classifier.train(train_tokens, y_train)

accuracy = ngram_classifier.evaluate(test_tokens, y_test)
print(f"4-Gram Model - Accuracy: {accuracy*100:.2f}%")
```
